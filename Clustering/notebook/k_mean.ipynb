# K-Means Algorithm for Clustering

## Hypothesis

K-Means is a clustering algorithm that aims to partition n observations into k clusters. There are three main steps in the algorithm:

### 1. Initialization: Randomly select k points as the initial centroids.

$$
u(0) = u1(0), u2(0), ..., uk(0)
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Generating a sample dataset
np.random.seed(42)
X = np.random.rand(10,2) * 10  # 10 points in 2D space

# Initial centroids randomly

# Initial centroids based on points A and B
C1 = np.array([3.75, 9.51])  # Centroid 1
C2 = np.array([7.32, 5.99])  # Centroid 2

# Plotting the data with initial centroids and color and color for each cluster

plt.scatter(X[:,0], X[:,1], s=100)
plt.scatter(C1, C2, s=100, c='r', marker='x')
plt.show()

```

### 2. Assignment: Assign each observation to the cluster with the nearest centroid.

$$
c(i) = argmin ||x(i) - u(j)(t)||^2
$$

where ***c(i)*** is the cluster assigned to observation ***i***, ***x(i)*** is the observation ***i***, ***u(j)*** is the centroid ***j***, and ***||.||*** is the Euclidean distance.

### 3. Update: Update the centroids to the mean of the observations in the cluster.


$$
u(j)(t+1) = (1/|S(j)|) * sum(x(i))
$$

where ***S(j)*** is the set of observations assigned to cluster ***j***.


```python
# Function to calculate Euclidean distance
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

def update_centroids(X, assignments):
    cluster1 = X[assignments == 0]
    cluster2 = X[assignments == 1]
    C1_new = np.mean(cluster1, axis=0) if cluster1.size else C1
    C2_new = np.mean(cluster2, axis=0) if cluster2.size else C2
    return C1_new, C2_new

# Function to assign each point to the closest centroid

def assign_points(X, C1, C2):
    assignments = []
    for point in X:
        if euclidean_distance(point, C1) < euclidean_distance(point, C2):
            assignments.append(0)
        else:
            assignments.append(1)
    return np.array(assignments)

# Function to plot the data with centroids and color for each cluster

def plot_clusters(X, C1, C2, assignments):
    cluster1 = X[assignments == 0]
    cluster2 = X[assignments == 1]
    plt.scatter(cluster1[:,0], cluster1[:,1], s=100)
    plt.scatter(cluster2[:,0], cluster2[:,1], s=100)
    plt.scatter(C1[0], C1[1], s=100, c='r', marker='x')
    plt.scatter(C2[0], C2[1], s=100, c='r', marker='x')
    plt.show()

# Main loop of the K-means algorithm
    
for i in range(10):
    assignments = assign_points(X, C1, C2)
    C1_new, C2_new = update_centroids(X, assignments)
    plot_clusters(X, C1, C2, assignments)
    print(C1, C2)

    #check if the centroids are not changing anymore

    # Check for convergence
    if np.allclose(C1, C1_new) and np.allclose(C2, C2_new):
        print(f"Converged after {i+1} iterations")
        break
    
    C1, C2 = C1_new, C2_new

# Final plot of the data with final centroids and color for each cluster

```
